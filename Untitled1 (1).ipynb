```python
1. What is a Random Variable in Probability Theory?
A random variable is a numerical value that represents the outcome of a random experiment.

It assigns numerical values to outcomes in a sample space.

2. What Are the Types of Random Variables?
Discrete random variables take countable values, such as the number of heads in coin tosses.

Continuous random variables take values from an infinite range, like height or temperature.

3. What is the Difference Between Discrete and Continuous Distributions?
Discrete distributions involve variables that have specific, countable values.

Continuous distributions involve variables that can take on any value within a given range.


4. What are Probability Distribution Functions (PDF)?
A PDF describes the likelihood of a random variable taking on a particular value.

For discrete variables, it gives the exact probability.

For continuous variables, it describes the density of probabilities, not actual probabilities.

5. How Do CDFs Differ from PDFs?
A cumulative distribution function (CDF) gives the probability that a random variable is less than or equal to a certain value.

Unlike the PDF, which shows probability at a specific point, the CDF shows cumulative probability up to a point.


6. What is a Discrete Uniform Distribution?
In a discrete uniform distribution, all outcomes have the same probability.

For example, when rolling a fair die, each face has an equal chance of appearing.

7. What Are the Key Properties of a Bernoulli Distribution?
A Bernoulli distribution has only two outcomes: success (1) or failure (0).

It is characterized by a single parameter, p, which is the probability of success.

8. What is the Binomial Distribution?
The binomial distribution models the number of successes in a fixed number of independent Bernoulli trials.

It is used when each trial has the same probability of success.

9. What is the Poisson Distribution and Where is It Applied?
The Poisson distribution models the number of events occurring in a fixed interval of time or space.

It is used for rare events, such as counting phone calls per hour or defects per unit length.


10. What is a Continuous Uniform Distribution?
In a continuous uniform distribution, all values within a certain interval are equally likely.

An example is generating a random number between 0 and 1.

11. What Are the Characteristics of a Normal Distribution?
A normal distribution is symmetric and bell-shaped.

Most values cluster around the mean, and probabilities decrease as values move away from the mean.

12. What is the Standard Normal Distribution?
The standard normal distribution is a normal distribution with a mean of 0 and a standard deviation of 1.

It is used to standardize values and compute Z-scores.


13. What is the Central Limit Theorem (CLT)?
The Central Limit Theorem states that the sampling distribution of the sample mean approaches a normal distribution as the sample size becomes large, regardless of the population distribution.

14. Why is CLT Important?
The CLT is important because it allows the use of normal distribution techniques for inference, even when the population distribution is not normal.

15. How Does the CLT Relate to the Normal Distribution?
The CLT explains why sample means tend to be normally distributed, making the normal distribution widely applicable in statistics.


16. What is the Application of Z Statistics in Hypothesis Testing?
Z statistics are used to determine how far a sample mean is from the population mean when the population standard deviation is known.

17. How to Calculate a Z-Score?
A Z-score is calculated by subtracting the population mean from a data point and dividing by the population standard deviation.

It shows how many standard deviations a value is from the mean.


18. What Are Point Estimates and Interval Estimates?
A point estimate is a single value used to estimate a population parameter.

An interval estimate provides a range of values within which the parameter likely lies, such as a confidence interval.

19. What is the Significance of Confidence Intervals?
Confidence intervals provide a range that is likely to contain the true population parameter, giving a measure of estimation reliability.

20. What is the Relationship Between a Z-Score and a Confidence Interval?
Z-scores are used to calculate the margin of error when constructing a confidence interval around a sample mean.

21. How Are Z-Scores Used to Compare Different Distributions?
Z-scores standardize different distributions to allow meaningful comparison of values from those distributions.


22. What Are the Assumptions for Applying the CLT?
The sample must be randomly selected.

Samples should be independent.

The sample size should be large, generally n â‰¥ 30, especially if the population is not normal.

23. What is the Concept of Expected Value?
The expected value is the long-run average outcome of a random variable over many repetitions of an experiment.

It is calculated by multiplying each possible value by its probability and summing the results.

24. How Does a Probability Distribution Relate to the Expected Outcome?
A probability distribution shows the likelihood of different outcomes, and the expected value represents the average of these outcomes, weighted by their probabilities.
```
