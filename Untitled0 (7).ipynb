```python
1. What is Hypothesis Testing in Statistics?

A process to make decisions or inferences about population parameters using sample data.
Involves formulating a null hypothesis (H₀) and an alternative hypothesis (H₁).

2. What is the Null Hypothesis (H₀)?

A statement of no effect or no difference.
It’s the hypothesis we try to disprove or reject.

3. Alternative Hypothesis (H₁)?

A statement indicating the presence of an effect or difference.
It’s what you want to prove.

4. What is the Significance Level (α)?

Threshold for rejecting the null hypothesis.
Common values: 0.05 (5%), 0.01, etc.

Determines the probability of a Type 1 error.

5. What Does a P-Value Represent?

The probability of observing the test results under H₀.
Smaller p-values indicate stronger evidence against H₀.

6. Interpretation of P-value

P < α → Reject H₀
P ≥ α → Fail to reject H₀

7. Type 1 and Type 2 Errors

Type I (α): Rejecting H₀ when it's true.
Type II (β): Failing to reject H₀ when it's false.

8. One-tailed vs Two-tailed Test

One-tailed: Tests for effect in one direction.
Two-tailed: Tests for effect in either direction.

9. What is a Z-test?

Used when population variance is known and n > 30.
Based on normal distribution.

10. How to Calculate Z-score?

z = (x̄ - μ) / (σ / √n)
Measures how far a sample mean is from the population mean.

11. What is the T-distribution?
Similar to normal but with fatter tails.

Used when sample size is small (n < 30) and population standard deviation is unknown.

12. Z-test vs T-test

Feature	Z-test	T-test
Known σ	Yes	No
Sample Size	Large (n > 30)	Small (n < 30)
Distribution	Normal	t-distribution

13. T-test in Hypothesis Testing

Compares means of:
A sample vs population (One-sample)
Two independent groups (Independent T-test)
Two related groups (Paired T-test)

14. Confidence Interval (CI)

Range of values likely to contain the population parameter.
CI = x̄ ± Z*(σ/√n) or x̄ ± t*(s/√n)

15. Margin of Error (MoE)

Half the width of a confidence interval.
Smaller MoE = more precise estimate.

16. Bayes’ Theorem in Statistics

Updates probability based on new evidence.
P(A|B) = [P(B|A) * P(A)] / P(B)
17. Chi-square Distribution
Used for categorical variables.

Tests:
Goodness of Fit
Independence

18. Chi-square Goodness of Fit

Compares observed vs expected frequencies in a single categorical variable.

19. F-distribution

Right-skewed, used for comparing variances.
Basis for ANOVA and F-tests.

20. ANOVA (Analysis of Variance)

Tests whether 3+ group means are significantly different.
Assumptions: independence, normality, equal variances.

21. Types of ANOVA

One-way ANOVA – One independent variable.
Two-way ANOVA – Two independent variables.

22. F-test

Compares variances of two populations.
Used in ANOVA and regression analysis.

```

```python
#2. Discrete Uniform Distribution – PMF

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 7)
pmf = [1/6] * 6
plt.stem(x, pmf, basefmt=" ")
plt.title("PMF of Discrete Uniform Distribution")
plt.show()
```

```python
#3. PDF of Bernoulli Distribution

def bernoulli_pmf(p, x):
    return p**x * (1-p)**(1-x)

print(bernoulli_pmf(0.7, 1))
```

```python
#4. Simulate Binomial Distribution & Plot Histogram

from scipy.stats import binom
data = binom.rvs(n=10, p=0.5, size=1000)
plt.hist(data, bins=10)
plt.title("Binomial Distribution")
plt.show()
```

```python
#5. Poisson Distribution Visualization

from scipy.stats import poisson
x = np.arange(0, 15)
pmf = poisson.pmf(x, mu=4)
plt.stem(x, pmf)
plt.title("Poisson Distribution")
plt.show()
```

```python
#6. CDF of Discrete Uniform Distribution

from scipy.stats import randint
x = np.arange(1, 7)
cdf = randint.cdf(x, 1, 7)
plt.step(x, cdf)
plt.title("CDF of Discrete Uniform")
plt.show()
```

```python
#7. Continuous Uniform Distribution

from scipy.stats import uniform
x = np.linspace(0, 1, 1000)
pdf = uniform.pdf(x)
plt.plot(x, pdf)
plt.title("Uniform Distribution (Continuous)")
plt.show()
```

```python
#8. Simulate Normal Distribution & Histogram

data = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(data, bins=30, density=True)
plt.title("Normal Distribution")
plt.show()
```

```python
#9. Calculate Z-scores & Plot

from scipy.stats import zscore
data = np.random.normal(50, 10, 100)
z_scores = zscore(data)
plt.hist(z_scores, bins=30)
plt.title("Z-scores")
plt.show()
```

```python
#10. Central Limit Theorem Implementation

means = [np.mean(np.random.exponential(scale=2, size=30)) for _ in range(1000)]
plt.hist(means, bins=30)
plt.title("CLT with Exponential Distribution")
plt.show()
```
